```python
import pandas as pd 
```

```python

df_jan = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet')

df_feb = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
```

```python
df_jan.info()
```

```python
df_jan['duration'] = df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime']
#df_jan['duration'].total_seconds() / 60
#int(timesince.total_seconds() / 60)

df_jan['duration']
```

```python
df_jan['duration'] = df_jan['duration'].dt.total_seconds() / 60
```

```python
df_jan.describe()
```

```python
(df_jan['duration'] <= 60) & (df_jan['duration'] >=0 )
```

```python
df_clean= df_jan[(df_jan['duration'] <= 60) & (df_jan['duration'] >=1 )]
```

```python
(len(df_clean)/len(df_jan)) 
```

```python
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse=False)
```

```python
df_clean
```

```python
dicts = df_clean[['PULocationID', 'DOLocationID']].astype(str).to_dict(orient='records')
oh_array = dv.fit_transform(dicts)
```

```python
oh_array.shape
```

```python
num_unique_pu = df_jan['PULocationID'].nunique()
num_unique_do = df_jan['DOLocationID'].nunique()
total_features = num_unique_pu + num_unique_do
total_features
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(oh_array,df_clean['duration'])
```

```python
from sklearn.metrics import root_mean_squared_error
y_pred = lr.predict(oh_array)

root_mean_squared_error(df_clean['duration'], y_pred)
```

```python
df_feb['duration'] = df_feb['tpep_dropoff_datetime'] - df_feb['tpep_pickup_datetime']
df_feb['duration'] = df_feb['duration'].dt.total_seconds() / 60

df_val= df_feb[(df_feb['duration'] <= 60) & (df_feb['duration'] >=1 )]
dicts = df_val[['PULocationID', 'DOLocationID']].astype(str).to_dict(orient='records')
val_array = dv.transform(dicts)
y_pred_val = lr.predict(val_array)
root_mean_squared_error(df_val['duration'], y_pred_val)
```
